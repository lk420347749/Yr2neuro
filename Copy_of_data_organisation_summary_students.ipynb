````python
import glob
import os
import re

def combine_summary_files(base_path="EXPT", output_file="combined_output.txt"):
    """
    Combines summary files from ML4 folders, adding a DIV column.
    """
    ml4_pattern = os.path.join(base_path, "*", "DIV*", "*_ML4")
    ml4_folders = glob.glob(ml4_pattern)
    ml4_folders = [folder for folder in ml4_folders if os.path.isdir(folder)]

    if not ml4_folders:
        print(f"Error: No folders matching '*_ML4' found within DIV folders under {base_path}.")
        return

    summary_file_paths = []
    for ml4_folder in ml4_folders:
        summary_file_path = os.path.join(ml4_folder, "eventer.output", "ALL_events", "summary.txt")
        if os.path.exists(summary_file_path):
            summary_file_paths.append(summary_file_path)
        else:
            print(f"Warning: summary.txt not found in {summary_file_path}")

    if not summary_file_paths:
        print("Error: No summary.txt files found in the matching ML4 folders.")
        return

    with open(output_file, 'w') as outfile:
        for file_path in summary_file_paths:  # Changed 'file' to 'file_path' for clarity
            # Extract DIV from file path using regular expression
            div_match = re.search(r'DIV\s+(\d+)', file_path, re.IGNORECASE)
            div_value = div_match.group(1) if div_match else "Unknown"

            try:
                with open(file_path, 'r') as infile:
                    # Write DIV line to output file
                    outfile.write(f"DIV: {div_value}\n")

                    # Write rest of the summary data
                    outfile.write(infile.read() + "\n")
            except FileNotFoundError:
                print(f"File {file_path} not found.")
            except Exception as e:
                print(f"An error occurred while reading {file_path}: {e}")

    print(f"Summary files have been combined into {output_file}")


if __name__ == "__main__":
    combine_summary_files()
    with open("combined_output.txt", 'r') as f:
        print(f.read())

````

````python
#PROTO-3 -> Prompt "Extracts data from all summary.txt files found within the root directory and its subdirectories."
#use CODE 1 as IMPROVEMENT



import os
import re

def extract_data_from_files(root_dir):
    """
    Extracts data from all summary.txt files found within the root directory and its subdirectories.

    Args:
        root_dir: The root directory to search for summary.txt files.

    Returns:
        A list of dictionaries, where each dictionary represents the data extracted from one summary.txt file.
    """
    data_list = []
    for dirpath, dirnames, filenames in os.walk(root_dir):
        for filename in filenames:
            if filename.lower() == 'summary.txt':  # Case-insensitive file name check
                file_path = os.path.join(dirpath, filename)
                with open(file_path, 'r') as file_handle:
                    raw_data = file_handle.read()
                    data_dict = extract_data_from_text(raw_data)

                    # Add DIV information to data_dict
                    div_match = re.search(r'DIV\s+(\d+)', dirpath, re.IGNORECASE)
                    if div_match:
                        data_dict['DIV'] = int(div_match.group(1))

                    else:
                        data_dict['DIV'] = None  # Handle cases where DIV is not found

                    data_list.append(data_dict)
    return data_list

def extract_data_from_text(text):
    """
    Extracts data from the raw text content of a summary.txt file.

    Args:
        text: The raw text content of the summary.txt file.

    Returns:
        A dictionary containing the extracted data.
    """
    data_dict = {}
    for line in text.split('\n'):
        if ':' in line:
            key, value = line.split(':', 1)
            key = key.strip()
            value = value.strip()
            try:
                # Try to convert value to float if possible
                value = float(value)
            except ValueError:
                pass  # Keep value as string if conversion fails
            data_dict[key] = value
    return data_dict

# Example usage:
root_directory = 'EXPT'
data = extract_data_from_files(root_directory)
print(data)


import os
import re
import pandas as pd

# ... (extract_data_from_files and extract_data_from_text functions remain the same) ...

# Example usage:
root_directory = 'EXPT'
data = extract_data_from_files(root_directory)

# Create a Pandas DataFrame
df = pd.DataFrame(data)

# Display the DataFrame as a table
display(df)
````
